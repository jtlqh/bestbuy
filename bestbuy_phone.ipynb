product_name has other information that need to split into 2 or more columns

```python

product=raw_df['product'].copy().drop_duplicates()
#print(product)
temp = product.str.split('with', expand=True)
temp.columns=['0','1']
#temp['0'].str.split('-',expand=True)
memory= temp['1'].apply(lambda x: re.findall(r'\d+GB', str(x))).apply(lambda x: x[0] if len(x)>0 else '')
temp
```

```python
product_df = temp['0'].str.split('-',expand=True).drop(2, axis=1)
product_df.columns = ['brand', 'product_name']
```

```python
product_df
```

```python
product_name = product_df['product_name'].apply(lambda x: ' '.join([s for s in x.split() if s.find('GB')<0]))
```

```python
product_name
```

```python
m = product_df['product_name'].apply(lambda x: x.split()[-1]).apply(lambda x: x if 'GB' in x else '')
```

```python
memory.loc[memory==''] = m.loc[memory=='']
memory
```

```python
product_df.loc['product_name']=product_name
product_df['memory']=memory
product_df['price']=raw_df['price']
```

Some of the price are actually the monthly subscription price, so I will remove these entries

```python
product_df = product_df.loc[product_df['price']>100].sort_values(by=['brand','product_name'])
```

```python
product_df.drop_duplicates(inplace=True)
product_df
```
